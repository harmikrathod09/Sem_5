# Step 21: Pltot Target Distribution

```python
target_counts = heart['target'].value_counts().sort_index()

plt.bar(target_counts.index.astype(str), target_counts.values, color='skyblue', edgecolor='black')
plt.title('Target Distribution')
plt.xlabel('Target')
plt.ylabel('Count')
plt.grid(True, axis='y')
plt.show()
```

# Step 22 : Compare age by heart disease

```python
age_0 = heart[heart['target'] == 0]['age']
age_1 = heart[heart['target'] == 1]['age']

plt.boxplot([age_0, age_1], labels=['No Disease (0)', 'Disease (1)'])

plt.title('Age vs Heart Disease')
plt.xlabel('Target')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

# Step 23 : Scatter Plot of age vs thalach colored by target

```python
colors = heart['target'].map({0: 'blue', 1: 'red'})

plt.scatter(heart['age'], heart['thalach'], c=colors, alpha=0.7, edgecolors='black')

plt.title('Age vs Max Heart Rate (Thalach) by Heart Disease')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate (Thalach)')
plt.grid(True)

import matplotlib.patches as mpatches
no_disease = mpatches.Patch(color='blue', label='No Disease')
disease = mpatches.Patch(color='red', label='Disease')
plt.legend(handles=[no_disease, disease])

plt.show()
```

# Step 24 : Bar Chart â€“ Chest Pain Type vs Heart Disease

```python
cp_target = heart.groupby(['cp', 'target']).size().unstack(fill_value=0)

cp_target.plot(kind='bar', figsize=(8, 5), color=['blue', 'red'], edgecolor='black')

plt.title('Chest Pain Type vs Heart Disease')
plt.xlabel('Chest Pain Type (cp)')
plt.ylabel('Count')
plt.legend(['No Disease', 'Disease'])
plt.grid(True)
plt.show()
```

# Step 25 : Line Plot â€“ Average Cholesterol by Age

```python
avg_chol_by_age = heart.groupby('age')['chol'].mean()

plt.plot(avg_chol_by_age.index, avg_chol_by_age.values, color='green', marker='o')

plt.title('Average Cholesterol by Age')
plt.xlabel('Age')
plt.ylabel('Average Cholesterol')
plt.grid(True)
plt.show()
```
