# Step 1: Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('OnlineRetail.csv', encoding='ISO-8859-1')
```

```python
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
drop_with_dropna=df.dropna(subset=["InvoiceNo","Description"])
```

```python
drop_with_dropna
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
remove_canceled_invoice=df[~df['InvoiceNo'].str.lower().str.startswith('c')]
```

```python
remove_canceled_invoice
```

# Step 4: Keep Only Positive Quantities

```python
positive_quantites=df[df['Quantity']>0]
```

```python
positive_quantites
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower().str.replace(' ', '', regex=False)
```

### df

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
single_country=df[df['Country']=='United Kingdom']
```

```python
single_country
```

# Step 7: Remove Duplicates

```python
remove_duplicate=df.drop_duplicates()
```

```python
remove_duplicate
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df['TotalPrice']=df['Quantity']*df['UnitPrice']
```

```python
df
```

# Step 9 : Check for Top 10 Selling Products

```python
top_10_products_df = (df.groupby('Description')['Quantity'].sum().reset_index().sort_values(by='Quantity', ascending=False).head(10))
print(top_10_products_df)
```

# Step 10 : Identify Most Active Customers

```python
top_customers_qty = (df.groupby('CustomerID')['Quantity'].sum().sort_values(ascending=False).head(10))
print(top_customers_qty)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Month'] = df['InvoiceDate'].dt.month
monthly_sales = df.groupby('Month')['TotalPrice'].sum()
plt.bar(monthly_sales.index, monthly_sales.values, color='steelblue')
plt.title('Monthly Sales Volume')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
